```python
import pandas as pd
```

```python
# read the population by age and sex table
popByAgeSex = pd.read_csv("Data/WPP2017_PopulationByAgeSex_Medium.csv")
popByAgeSex.head()
```

```python
# read the locations table
locations = pd.read_csv("DataDocumentation/WPP2017_F01_LOCATIONS.csv")
locations.head()
```

```python
# Do the merge
combinedData = popByAgeSex.merge(locations, left_on='Location', right_on='RegionSubregionCountryArea', how='left')
combinedData.head()
```

```python
# Check number of rows
for df in [popByAgeSex, locations, combinedData]:
    print('The shape is {}'.format(df.shape))
```

```python
# Super, shapes are as expected! Now to write to a csv file.
combinedData.to_csv(r'Data/CombinedData.csv')
```

```python
# FERTILITY RATE
# Stack the columns
# remove the series name column
fertility = fertility.drop(columns=['Series Name', 'Series Code'])
# set indices 
idx = ['Country Name', 'Country Code']
fertility = fertility.set_index(idx)
# stack years
fertility = fertility.stack(dropna=False).reset_index()
# rename columns
fertility.columns=['Country Name', 'Country Code', 'Year', 'FertilityRate']
fertility['Year'] = pd.to_numeric(fertility['Year'])
# show results
fertility.head()
```
